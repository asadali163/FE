```python
sell_by_shop_sr = sell_out_sr.groupby("Outlet SF ID")["Volume in Packs"].sum()
# sell_out_sr.head()
```

```python
sell_by_shop_hr = df_sellout.groupby("customer_code")["sales_quantity"].sum()
```

```python
result = pd.DataFrame({"sell_sr": sell_by_shop_sr, "sell_hr": sell_by_shop_hr})

# Check equality
result["is_equal"] = result["sell_sr"] == result["sell_hr"]

# Difference (sr - hr)
result["difference"] = result["sell_sr"] - result["sell_hr"]
```

```python
result.sort_values("difference", ascending=False).head()
```
